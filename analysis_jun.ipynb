```python
dayton = sssm_combine[sssm_combine['region']== 'Dayton, OH']
dayton
```

```python
dayton.plot(kind='line', x='date', y='value')
```

```python
# getting the listing for Dayton, OH
dayton_inventory = issm[issm['region_id'].isin([845158, 394521])]
dayton_inventory.sort_values(by='date')
```

```python
# combining price and listing
dayton_sssm_issm = pd.merge(dayton, dayton_inventory, on='date',suffixes=['_price','_listing'])
dayton_sssm_issm
```

```python
# keep only those rows with the same region_id
dayton_sssm_issm_filter = dayton_sssm_issm[dayton_sssm_issm['region_id_price'] == dayton_sssm_issm['region_id_listing']]
dayton_sssm_issm_filter
```

```python
# keep only those rows needed for a scatter plot
dayton_sssm_issm_filter_scatter= dayton_sssm_issm_filter[['date','value_price','value_listing']]
dayton_sssm_issm_filter_scatter
```

```python
# get the linear regression parameters
from scipy import stats
(slope, intercept, rvalue, pval, stderr) = stats.linregress(dayton_sssm_issm_filter_scatter['value_listing'],\
                                                            dayton_sssm_issm_filter_scatter['value_price'])
```

```python
# plot the scatter plot and regression line.
import matplotlib.pyplot as plt
dayton_sssm_issm_filter_scatter.plot(kind='scatter',x='value_listing', y='value_price', title='house price vs the number of listing')
regr_yval = dayton_sssm_issm_filter_scatter['value_listing']*slope + intercept
plt.plot(dayton_sssm_issm_filter_scatter['value_listing'], regr_yval, color='r')
print(f'r squared value is {rvalue**2:.5f}')
```

From the graph we can we that there is generally a negative correlation between the house price and number of listing. This makes sense from a supply-demand perspective. As the more supply (listing) there is, the lower the price will be given the same demand. The r-squared value is significant, meaning the regression model is reliable.
